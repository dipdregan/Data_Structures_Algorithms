##  Recursion
**when a fuction call itself until a specifical condition is met**


```python
def f():
    print('hello')
    return f()
    

#calling the fuction
f()
```

```python
def f1():
    print('hello')
    f()
if __name__=="__main__":
    f()
```

```python
count = 0

def f():
    count
    print(f'hi...this is your {count} count')
    count = count + 1
    f()

f()
```

```python
count = 0

def f():
    global count  # Declare 'count' as a global variable
    print(f'hi...this is your {count} count')
    count = count + 1
    f()

f()
```

```python
count  = 1
def f():
    global count
    if count ==10:
        return
    print(count)
    count +=1
    f()

def f1():
    print('your fuction f() rained complately,,.')
    
if __name__ == "__main__":
    f()
    f1()
```

### Basic Recursion Problem

```python
#Q1. Print Name 5 times using recursion

count = 0

def print_name(name):
    global count
    if count == 5:
        return
    print(name)
    count += 1
    print_name(name)

print_name('dipak')
```

```python
# Q2 . print linearly from 1 to n
def print_no(n):
    if n == 1:
        print(n)
        return
    print_no(n - 1)
    print(n)

print_no(10)
```

```python
# Q2 . print linearly from 1 to n
count = 0
def print_no(n):
    global count
    if count==n:
        return
    count+=1
    print(count)
    print_no(n)
print_no(5)
```

```python
# Q3. print linearly 1 to n
def print_1_n(n):
    count = n
    if count==1:
        return
    print(count)
    count-=1
    
    print_1_n(count)
print_1_n(10)
```
