```python
class Solution:
    def isPowerOfThree(self,n):
        if n == 1:
            return True
        elif n <= 0 or n % 3 != 0:
            return False
        else:
            return self.isPowerOfThree(n // 3)

```

```python
def run_test_case(test_case_number: int, N: int, expected_output: bool) -> None:
    output = Solution().isPowerOfThree(N)
    status = "Pass" if output == expected_output else "Failed"
    
    print(f"Test Case {test_case_number}:\n{'==' * 10}")
    print(f"Input: N = {N}")
    print(f"Output: {output}")
    print(f"Expected Output: {expected_output}")
    print(f"Status: {status}\n")

    
N1 = 27
expected_output1 = True
run_test_case(1, N1, expected_output1)

N2 = 0
expected_output2 = False
run_test_case(2, N2, expected_output2)

N3 = -1
expected_output3 = False
run_test_case(3, N3, expected_output3)
```

```python
class Solution:
    def lastRemaining(self, n: int) -> int:
        arr = list(range(1, n+1))
        left_to_right = True

        while len(arr) > 1:
            if left_to_right:
                arr = arr[1::2]
            else:
                arr = arr[-2::-2][::-1]
            left_to_right = not left_to_right

        return arr[0]
```

```python
def run_test_case(test_case_number: int, n: int, expected_output: int) -> None:
    output = Solution().lastRemaining(n)
    status = "Pass" if output == expected_output else "Failed"

    print(f"Test Case {test_case_number}:\n{'==' * 10}")
    print(f"Input: n = {n}")
    print(f"Output: {output}")
    print(f"Expected Output: {expected_output}")
    print(f"Status: {status}\n")


# Test Case 1
n1 = 9
expected_output1 = 6
run_test_case(1, n1, expected_output1)

# Test Case 2
n2 = 1
expected_output2 = 1
run_test_case(2, n2, expected_output2)
```

```python
class Solution:
    def subsets(self, string: str):
        subsets = []
        self.generate_subsets(string, 0, "", subsets)
        return subsets

    def generate_subsets(self, string, index, current_subset, subsets):
        if index == len(string):
            subsets.append(current_subset)
            return

        self.generate_subsets(string, index + 1, current_subset, subsets)
        self.generate_subsets(string, index + 1, current_subset + string[index], subsets)
```

```python
def run_test_case(test_case_number: int, string: str, expected_output: list) -> None:
    output = Solution().subsets(string)
    output.sort()
    expected_output.sort()
    status = "Pass" if output == expected_output else "Failed"

    print(f"Test Case {test_case_number}:\n{'==' * 10}")
    print(f"Input: string = {string}")
    print(f"Output: {output}")
    print(f"Expected Output: {expected_output}")
    print(f"Status: {status}\n")


# Test Case 1
string1 = "abc"
expected_output1 = ["", "a", "ab", "abc", "ac", "b", "bc", "c"]
run_test_case(1, string1, expected_output1)

# Test Case 2
string2 = "abcd"
expected_output2 = ["", "a", "ab", "abc", "abcd", "abd", "ac", "acd", "ad", "b", "bc", "bcd", "bd", "c", "cd", "d"]
run_test_case(2, string2,expected_output2)
```

```python
class Solution:
    def calculateLength(self, string):
        if string == "":
            return 0
        else:
            return 1 + self.calculateLength(string[1:])
```

```python
class Solution:
    def countSubstrings(self, S):
        count = 0
        self.count_substrings_helper(S, 0, len(S) - 1, count)
        return count

    def count_substrings_helper(self, S, start, end, count):
        if start > end:
            return
        
        if S[start] == S[end]:
            count += 1

        self.count_substrings_helper(S, start + 1, end, count)
        self.count_substrings_helper(S, start, end - 1, count)
```

```python
def run_test_case(test_case_number: int, S: str, expected_output: int) -> None:
    output = Solution().countSubstrings(S)
    status = "Pass" if output == expected_output else "Failed"

    print(f"Test Case {test_case_number}:\n{'==' * 10}")
    print(f"Input: S = '{S}'")
    print(f"Output: {output}")
    print(f"Expected Output: {expected_output}")
    print(f"Status: {status}\n")


# Test Case 1
S1 = "abcab"
expected_output1 = 7
run_test_case(1, S1, expected_output1)

# Test Case 2
S2 = "aba"
expected_output2 = 4
run_test_case(2, S2, expected_output2)
```

```python
class Solution:
    def towerOfHanoi(self, N):
        moves = []
        self.move_disks(N, 1, 3, 2, moves)
        return moves, len(moves)

    def move_disks(self, N, source, destination, auxiliary, moves):
        if N == 1:
            moves.append(f"move disk 1 from rod {source} to rod {destination}")
            return

        self.move_disks(N - 1, source, auxiliary, destination, moves)
        moves.append(f"move disk {N} from rod {source} to rod {destination}")
        self.move_disks(N - 1, auxiliary, destination, source, moves)
```

```python
def run_test_case(test_case_number: int, N: int, expected_output: tuple) -> None:
    moves, total_moves = Solution().towerOfHanoi(N)
    expected_moves, expected_total_moves = expected_output
    status = "Pass" if (moves == expected_moves) and (total_moves == expected_total_moves) else "Failed"

    print(f"Test Case {test_case_number}:\n{'==' * 10}")
    print(f"Input: N = {N}")
    print(f"Output - Moves: {moves}")
    print(f"Output - Total Moves: {total_moves}")
    print(f"Expected Output - Moves: {expected_moves}")
    print(f"Expected Output - Total Moves: {expected_total_moves}")
    print(f"Status: {status}\n")


# Test Case 1
N1 = 2
expected_output1 = (["move disk 1 from rod 1 to rod 2", "move disk 2 from rod 1 to rod 3", "move disk 1 from rod 2 to rod 3"], 3)
run_test_case(1, N1, expected_output1)

# Test Case 2
N2 = 3
expected_output2 = (["move disk 1 from rod 1 to rod 3", "move disk 2 from rod 1 to rod 2", "move disk 1 from rod 3 to rod 2", 
                    "move disk 3 from rod 1 to rod 3", "move disk 1 from rod 2 to rod 1", "move disk 2 from rod 2 to rod 3",
                    "move disk 1 from rod 1 to rod 3"], 7)
run_test_case(2, N2, expected_output2)
```

```python
from typing import List

class Solution:
    def get_permutations(self, string):
        if len(string) <= 1:
            return [string]

        permutations = []
        for i in range(len(string)):
            fixed_char = string[i]
            remaining_chars = string[:i] + string[i+1:]
            sub_permutations = self.get_permutations(remaining_chars)

            for sub_permutation in sub_permutations:
                permutations.append(fixed_char + sub_permutation)

        return permutations
```

```python
def run_test_case(test_case_number: int, input_string: str, expected_output: List[str]) -> None:
    output = Solution().get_permutations(input_string)
    status = "Pass" if output == expected_output else "Failed"

    print(f"Test Case {test_case_number}:\n{'==' * 10}")
    print(f"Input: input_string = {input_string}")
    print(f"Output: {output}")
    print(f"Expected Output: {expected_output}")
    print(f"Status: {status}\n")


# Test Case 1
input_string1 = "cd"
expected_output1 = ["cd", "dc"]
run_test_case(1, input_string1, expected_output1)

# Test Case 2
input_string2 = "abb"
expected_output2 = ["abb", "bab", "bba"]
run_test_case(2, input_string2, expected_output2)
```

```python
class Solution:
    def count_consonants(self, string):
        vowels = 'aeiouAEIOU'

        if string == "":
            return 0
        elif string[0].isalpha() and string[0] not in vowels:
            return 1 + self.count_consonants(string[1:])
        else:
            return self.count_consonants(string[1:])
```

```python

def run_test_case(test_case_number: int, string: str, expected_output: int) -> None:
    output = Solution().count_consonants(string)
    status = "Pass" if output == expected_output else "Failed"

    print(f"Test Case {test_case_number}:\n{'==' * 10}")
    print(f"Input: string = {string}")
    print(f"Output: {output}")
    print(f"Expected Output: {expected_output}")
    print(f"Status: {status}\n")


string1 = "abc de"
expected_output1 = 3
run_test_case(1, string1, expected_output1)

string2 = "geeksforgeeks portal"
expected_output2 = 12
run_test_case(2, string2, expected_output2)
```
