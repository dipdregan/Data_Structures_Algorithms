```python
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        if n == 1:
            return True
        elif n <= 0 or n % 2 != 0:
            return False
        else:
            return self.isPowerOfTwo(n // 2)
```

```python
def run_test_case(test_case_number: int, n: int, expected_output: bool) -> None:
    solution = Solution()
    output = solution.isPowerOfTwo(n)
    status = "Pass" if output == expected_output else "Failed"
    print(f"Test Case {test_case_number} -\n{'==' * 10}\nInput: {n}\nOutput: {output}\nExpected Output: {expected_output}\nStatus: {status}\n")

run_test_case(1, 1, True)
run_test_case(2, 16, True)
run_test_case(3, 3, False)
```

```python
class Solution:
    def power_of_number(self, N, P):
        if P == 0:
            return 1
        elif P == 1:
            return N
        else:
            return N * self.power_of_number(N, P - 1)

```

```python
def run_test_case(test_case_number: int, N: int, P: int, expected_output: int) -> None:
    output = Solution().power_of_number(N, P)
    status = "Pass" if output == expected_output else "Failed"
    print(f"Test Case {test_case_number} -\n{'==' * 10}\nInput: N = {N}, P = {P}\nOutput: {output}\nExpected Output: {expected_output}\nStatus: {status}\n")

N1 = 5
P1 = 2
expected_output1 = 25
run_test_case(1, N1, P1, expected_output1)

N2 = 2
P2 = 5
expected_output2 = 32
run_test_case(2, N2, P2, expected_output2)
```

```python
class Solution:
    def find_maximum(self, arr):
        if len(arr) == 1:
            return arr[0]
        else:
            return max(arr[0], self.find_maximum(arr[1:]))
```

```python
def run_test_case(test_case_number: int, arr: list, expected_output: int) -> None:
    output = Solution().find_maximum(arr)
    status = "Pass" if output == expected_output else "Failed"
    print(f"Test Case {test_case_number} -\n{'==' * 10}\nInput: arr = {arr}\nOutput: {output}\nExpected Output: {expected_output}\nStatus: {status}\n")

arr1 = [1, 4, 3, -5, -4, 8, 6]
expected_output1 = 8
run_test_case(1, arr1, expected_output1)

arr2 = [1, 4, 45, 6, 10, -8]
expected_output2 = 45
run_test_case(2, arr2, expected_output2)
```

```python
class Solution:
    def find_nth_term(self, a, d, N):
        if N == 1:
            return a
        else:
            return self.find_nth_term(a + d, d, N - 1)

```

```python
def run_test_case(test_case_number: int, a: int, d: int, N: int, expected_output: int) -> None:
    output = Solution().find_nth_term(a, d, N)
    status = "Pass" if output == expected_output else "Failed"
    print(f"Test Case {test_case_number} -\n{'==' * 10}\nInput: a = {a}, d = {d}, N = {N}\nOutput: {output}\nExpected Output: {expected_output}\nStatus: {status}\n")

a1 = 2
d1 = 1
N1 = 5
expected_output1 = 6
run_test_case(1, a1, d1, N1, expected_output1)

a2 = 5
d2 = 2
N2 = 10
expected_output2 = 23
run_test_case(2, a2, d2, N2, expected_output2)
```

```python
class Solution:
    def print_permutations(self, s: str) -> None:
        def generate_permutations(s, current_permutation):
            if len(s) == 0:
                print(current_permutation)
                return

            for i in range(len(s)):
                remaining_chars = s[:i] + s[i+1:]
                generate_permutations(remaining_chars, current_permutation + s[i])

        generate_permutations(s, "")
```

```python

def run_test_case(test_case_number: int, s: str, expected_output: list) -> None:
    print(f"Test Case {test_case_number} -\n{'==' * 10}\nInput: S = {s}")
    print("Expected Output:")
    for perm in expected_output:
        print(perm)
    print("Output:")
    Solution().print_permutations(s)
    print()

S1 = "ABC"
expected_output1 = ["ABC", "ACB", "BAC", "BCA", "CAB", "CBA"]
run_test_case(1, S1, expected_output1)

S2 = "XY"
expected_output2 = ["XY", "YX"]
run_test_case(2, S2, expected_output2)

```
